```python
import requests
import json
from pathlib import Path
from calendar import monthrange
import pandas as pd
```

```python
base_url = "https://apidatos.ree.es/es/datos/balance/balance-electrico"

headers = {
    "Accept": "application/json",
    "Content-Type": "application/json"
}

years = [2024, 2025]

output_path = Path("../data/bronze/esios/ree_raw_2024_2025.json")

all_data = []
```

```python
for year in years:
    for month in range(1, 13):
        last_day = monthrange(year, month)[1]

        params = {
            "start_date": f"{year}-{month:02d}-01T00:00",
            "end_date": f"{year}-{month:02d}-{last_day:02d}T23:59",
            "time_trunc": "day"
        }



        response = requests.get(base_url, headers=headers, params=params, timeout=30)
        response.raise_for_status()

        data = response.json()

        all_data.append({
            "year": year,
            "month": month,
            "data": data
        })
```

```python
with open(output_path, "w", encoding="utf-8") as f:
    json.dump(all_data, f, ensure_ascii=False, indent=2)

print("\nJSON conjunto guardado en:", output_path)
```
