```python
import boto3
import json
import pandas as pd
from io import BytesIO
from pathlib import Path
```

```python
bucket_name = "proyecto-electricidad-ela"   
s3_key = "bronze/calendario/festivos_2024_2025.json"

s3 = boto3.client("s3")

silver_output = Path("../data/silver/festivos/festivos.parquet")
silver_output.parent.mkdir(parents=True, exist_ok=True)
```

```python
response = s3.get_object(Bucket=bucket_name, Key=s3_key)
raw_data = response["Body"].read()

data = json.loads(raw_data.decode("utf-8"))

type(data), len(data)
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
df = df[["date", "localName", "global", "types", "year"]].copy()

# Renombrar columnas
df.rename(columns={
    "date": "fecha",
    "localName": "festivo_nombre",
    "global": "festivo_nacional",
    "types": "tipo_festivo"
}, inplace=True)

# Tipos de datos
df["fecha"] = pd.to_datetime(df["fecha"])
df["festivo_nacional"] = df["festivo_nacional"].astype(int)

# Variable clave
df["es_festivo"] = 1

# Normalizar tipo_festivo
df["tipo_festivo"] = df["tipo_festivo"].apply(
    lambda x: x[0] if isinstance(x, list) and len(x) > 0 else None
)

# Columnas derivadas
df["mes"] = df["fecha"].dt.month
df["dia"] = df["fecha"].dt.day

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.to_parquet(silver_output, index=False)
```
