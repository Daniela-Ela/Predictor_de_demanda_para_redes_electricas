```python
import json
import requests
from pathlib import Path
import pandas as pd
```

```python
YEARS = [2024, 2025]
COUNTRY = "ES"

output_dir = Path("../data/bronze/festivos")   
output_file = output_dir / "festivos_2024_2025.json"

all_data = []
```

```python
for year in YEARS:
    url = f"https://date.nager.at/api/v3/PublicHolidays/{year}/{COUNTRY}"
    
    response = requests.get(url, timeout=30)
    response.raise_for_status()
    
    data = response.json()
    
    for item in data:
        item["year"] = year
    
    all_data.extend(data)
```

```python
with open(output_file, "w", encoding="utf-8") as f:
    json.dump(all_data, f, ensure_ascii=False, indent=2)

print(f"JSON combinado guardado en: {output_file}")
```

```python
df_all = pd.DataFrame(all_data)
display(df_all.head())

print("\nAños presentes:")
print(df_all["year"].value_counts().sort_index())
```
