```python
import pandas as pd
import os

ruta = "../data/bronze/clima/enero2025/"
archivos = sorted(os.listdir(ruta))

lista = []

for archivo in archivos:
    if archivo.endswith(".xls"):
        df = pd.read_excel(ruta + archivo, header=4)

        fecha = archivo.replace("Aemet", "").replace(".xls", "")

        df = df[[
            "Provincia",
            "Temperatura media (ºC)",
            "Precipitación 00-24h (mm)"
        ]].copy()

        df["fecha"] = fecha

        lista.append(df)
```

```python
df_clima = pd.concat(lista, ignore_index=True)
```

```python
df_clima.columns = ["provincia", "temp_media", "precipitacion", "fecha"]
```

```python
df_clima.head()
```

```python
df_clima["fecha"] = pd.to_datetime(df_clima["fecha"])
```

```python
df_clima.dtypes
```

```python
df_clima.to_csv("../data/bronze/clima_enero_2025_unido.csv", index=False)
```

```python
df_clima.to_parquet("../data/silver/clima_enero_2025.parquet", index=False)
```
